**4)Answer the following questions with statistical evidence.**

**a)Do charges of people who smoke differ significantly from the people who 
don't?**

```python
#Formulating Hypothesis
Ha="Charges of people who smoke and who dont are not the same"
Ho="Charges of people who smoke and who dont are the same"
```

```python
#Since we need to compare the mean value of two different groups, so we will use t-test of mean
```

```python
from scipy.stats import ttest_ind
```

```python
x=np.array(df[df['smoker']=='yes'].charges)  #Extracting charges of smokers
y=np.array(df[df['smoker']=='no'].charges) #Extracting charges of non-smokers
```

```python
test,p_value=ttest_ind(x,y)
```

```python
test,p_value
```

```python
if p_value<0.05:
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

* CHarges are the **same** for both people who smoke and do not smoke.

**b) Does bmi of males differ significantly from females?**

```python
#Formulating hypothesis
Ha="Gender has a impact on bmi"
Ho="Gender does not have a impact on bmi"
```

```python
#Again using t-test of mean
```

```python
x=np.array(df[df['sex']=='female'].bmi)  #Extracting bmi of female and storing in a array
y=np.array(df[df['sex']=='male'].bmi)  #Extracting bmi of male and storing in a array
```

```python
test,p_value=ttest_ind(x,y)
```

```python
if p_value<0.05:
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

* Hence,gender **does not** have a impact on bmi.

**c) Is the proportion of smokers significantly different in different genders?**

```python
#Formulating hypothesis
Ha="Genders has an affect on smoking habits"
Ho="Genders has no affect on smoking habits"
```

```python
#Here we need to compare two categorical variables so we will use Chi Square test
```

```python
import scipy.stats as stats
```

```python
crosstab=pd.crosstab(df['sex'],df['smoker'])   #Formulating data in tabular format
```

```python
test,p_value,a,b=stats.chi2_contingency(crosstab)
```

```python
if p_value<0.05:
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

* Genders **has an affect** on smoking habits.

**d) Is the distribution of bmi across women with no children, one child and two children the same ?**

```python
#Formulating hypothesis
Ha="No. of children has affect on bmi"
Ho="No. of children has no affect on bmi"
```

```python
#Using Anova test
```

```python
female=df[df['sex']=='female']  #Extracting all the female data
```

```python
zero=female[female['children']==0]['bmi']
one=female[female['children']==1]['bmi']
two=female[female['children']==2]['bmi']
```

```python
test,p_value=stats.f_oneway(zero,one,two)
```

```python
if p_value<0.05:
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

* Number of children has **no affect** on bmi.
